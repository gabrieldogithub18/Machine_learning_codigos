```python
# definir um treshold minímino e um treshold maximo em quantidade_dos_alimentos para impedir que um certo alimento não
# ultrapasse a quantidade desejada ou não fique abixo de uma quantidade desejada
```

```python
class produtor_dieta:
    def __init__(self, array, quantidade_desejada, valores_de_certa_comida, nomes_alimentos):
        self.nomes_alimentos = nomes_alimentos
        self.array = array
        self.quantidade_desejada = quantidade_desejada
        self.valores_de_certa_comida = valores_de_certa_comida 
        self.n_nutrientes = len(quantidade_desejada[0])
            
    def somando_os_nutrientes_respectivos(self, quantidade_dos_alimentos):
        soma_dos_nutrientes_totais = np.zeros(shape= (self.n_nutrientes))
        for n in range(quantidade_dos_alimentos.shape[1]):
            soma_dos_nutrientes_totais = soma_dos_nutrientes_totais + quantidade_dos_alimentos[0][n] * (self.valores_de_certa_comida[n] / 100)
        return soma_dos_nutrientes_totais
    
    def multiplicar_K_por_respectivos_valores_valores(self, y_true, y_pred):
        somas_dos_nutrientes_respectivos = self.somando_os_nutrientes_respectivos(y_pred)
        return K.square(y_true - somas_dos_nutrientes_respectivos)
    
    def fazer_rede(self):
        entrada_gerador = Input(shape= (self.array.shape[1] ,) ) 
        x = entrada_gerador
        n_camadas_gerador = 8
        for n in range(n_camadas_gerador):
            if n < n_camadas_gerador - 1:
                camada = Dense(units= 15, activation= None)
            else:
                camada = Dense(units= self.array.shape[1], activation= 'relu')
            x = camada(x)

        saida_gerador = x
        self.gerador = Model(entrada_gerador, saida_gerador)
        self.gerador.compile(loss= self.multiplicar_K_por_respectivos_valores_valores, optimizer= 'adam' )
    
    def ver_summario(self):
        self.gerador.summary()
    
    def super_ajustar(self):
        self.gerador.fit( self.array , self.quantidade_desejada , epochs= 500 )
        
    def predizer(self):
        self.q_alimento = self.gerador.predict(self.array)
        self.nomes_quantidade = dict(zip(self.nomes_alimentos, self.q_alimento[0]))
        
    def ver_dif_exp_pred(self):
        self.nutricao_total_q_alimento = self.somando_os_nutrientes_respectivos(self.q_alimento)
        print(self.quantidade_desejada - self.nutricao_total_q_alimento)
```
